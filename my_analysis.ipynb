```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://huggingface.co/datasets/maharshipandya/spotify-tracks-dataset/resolve/main/dataset.csv"
df = pd.read_csv(url)

df = df.head(101)

df.to_csv("music_data.csv", index=False)

from google.colab import files
files.download("music_data.csv")

df.isna().sum()
df = df.drop_duplicates()
df = df.dropna()

df.head()

plt.figure(figsize=(10, 5))
plt.hist(df["tempo"], bins=40)
plt.title("Distribution of Song Tempos (BPM)")
plt.xlabel("Tempo (BPM)")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(10, 5))
plt.scatter(df["energy"], df["loudness"], alpha=0.3)
plt.title("Energy vs Loudness in Songs")
plt.xlabel("Energy")
plt.ylabel("Loudness (dB)")
plt.show()
```
